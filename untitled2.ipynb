```python
# Clasificador de Exoplanetas - VERSIÓN OPTIMIZADA CON MAPEO DE FEATURES

# 1. IMPORTACIÓN DE LIBRERÍAS

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.impute import SimpleImputer
import warnings
warnings.filterwarnings('ignore')
```

```python
# 2. CARGA DE DATOS

from google.colab import files
print("Sube tus 7 archivos CSV:")
uploaded = files.upload()

def load_csv_flexible(filename):
    configs = [
        {'sep': ',', 'comment': '#', 'skipinitialspace': True, 'low_memory': False},
        {'sep': ',', 'comment': '#', 'skipinitialspace': True, 'encoding': 'latin-1', 'low_memory': False},
    ]
    for config in configs:
        try:
            df = pd.read_csv(filename, **config)
            if len(df) > 0:
                return df
        except:
            continue
    return None

def load_data():
    archivos = [
        ('candidato_kepler.csv', 'CANDIDATO', 'Kepler'),
        ('candidato_tess.csv', 'CANDIDATO', 'TESS'),
        ('candidato_k2.csv', 'CANDIDATO', 'K2'),
        ('falso_positivo_kepler.csv', 'FALSO_POSITIVO', 'Kepler'),
        ('falso_positivo_tess.csv', 'FALSO_POSITIVO', 'TESS'),
        ('falso_positivo_k2.csv', 'FALSO_POSITIVO', 'K2'),
        ('confirmados.csv', 'CONFIRMADO', 'Confirmado')
    ]

    df_list = []
    for filename, clase, origen in archivos:
        df = load_csv_flexible(filename)
        if df is not None:
            df['clase'] = clase
            df['origen'] = origen
            df_list.append(df)
            print(f"✓ {filename}: {len(df)} filas")

    return df_list

print("\n" + "="*70)
print("CARGANDO DATOS...")
print("="*70)
df_list = load_data()
```

```python
# 3. MAPEO INTELIGENTE DE FEATURES

print("\n" + "="*70)
print("MAPEO DE FEATURES ENTRE MISIONES")
print("="*70)

# Definir mapeo de columnas entre misiones
MAPEO_FEATURES = {
    # PARÁMETROS PLANETARIOS (ordenados por importancia)
    'Periodo_Orbital_dias': {
        'Kepler': 'koi_period',
        'TESS': 'pl_orbper',
        'K2': 'pl_orbper',
        'Confirmado': 'pl_orbper'
    },
    'Radio_Planeta_RE': {
        'Kepler': 'koi_prad',
        'TESS': 'pl_rade',
        'K2': 'pl_rade',
        'Confirmado': 'pl_rade'
    },
    'Profundidad_Transito_ppm': {
        'Kepler': 'koi_depth',
        'TESS': 'pl_trandep',
        'K2': None,  # No disponible en K2
        'Confirmado': None
    },
    'Duracion_Transito_hrs': {
        'Kepler': 'koi_duration',
        'TESS': 'pl_trandurh',
        'K2': None,
        'Confirmado': None
    },
    'Temp_Equilibrio_K': {
        'Kepler': 'koi_teq',
        'TESS': 'pl_eqt',
        'K2': 'pl_eqt',
        'Confirmado': 'pl_eqt'
    },
    'Flujo_Insolacion_FE': {
        'Kepler': 'koi_insol',
        'TESS': 'pl_insol',
        'K2': 'pl_insol',
        'Confirmado': 'pl_insol'
    },
    'Parametro_Impacto': {
        'Kepler': 'koi_impact',
        'TESS': None,
        'K2': None,
        'Confirmado': None
    },
    'SNR_Transito': {
        'Kepler': 'koi_model_snr',
        'TESS': None,
        'K2': None,
        'Confirmado': None
    },

    # PARÁMETROS ESTELARES
    'Teff_Estelar_K': {
        'Kepler': 'koi_steff',
        'TESS': 'st_teff',
        'K2': 'st_teff',
        'Confirmado': 'st_teff'
    },
    'Logg_Estelar': {
        'Kepler': 'koi_slogg',
        'TESS': 'st_logg',
        'K2': 'st_logg',
        'Confirmado': 'st_logg'
    },
    'Radio_Estelar_RS': {
        'Kepler': 'koi_srad',
        'TESS': 'st_rad',
        'K2': 'st_rad',
        'Confirmado': 'st_rad'
    },
    'Masa_Estelar_MS': {
        'Kepler': None,
        'TESS': None,
        'K2': 'st_mass',
        'Confirmado': 'st_mass'
    },
    'Metalicidad_Estelar': {
        'Kepler': None,
        'TESS': None,
        'K2': 'st_met',
        'Confirmado': 'st_met'
    },
    'Distancia_Sistema_pc': {
        'Kepler': None,
        'TESS': 'st_dist',
        'K2': 'sy_dist',
        'Confirmado': 'sy_dist'
    },

    # PARÁMETROS ORBITALES ADICIONALES
    'Semieje_Mayor_AU': {
        'Kepler': None,
        'TESS': None,
        'K2': 'pl_orbsmax',
        'Confirmado': 'pl_orbsmax'
    },
    'Excentricidad': {
        'Kepler': None,
        'TESS': None,
        'K2': 'pl_orbeccen',
        'Confirmado': 'pl_orbeccen'
    },

    # MASA PLANETARIA (muy importante si está disponible)
    'Masa_Planeta_ME': {
        'Kepler': None,
        'TESS': None,
        'K2': 'pl_bmasse',
        'Confirmado': 'pl_bmasse'
    }
}

# FLAGS IMPORTANTES DE KEPLER (para detectar falsos positivos)
FLAGS_KEPLER = {
    'Flag_No_Transito': 'koi_fpflag_nt',
    'Flag_Eclipse_Estelar': 'koi_fpflag_ss',
    'Flag_Offset_Centroide': 'koi_fpflag_co',
    'Flag_Contaminacion': 'koi_fpflag_ec',
    'Score_Disposicion': 'koi_score'
}

print("\n✓ Features mapeadas: " + str(len(MAPEO_FEATURES)))
print("✓ Flags de Kepler: " + str(len(FLAGS_KEPLER)))
```

```python
# 4. FUNCIÓN DE UNIFICACIÓN CON MAPEO

def unificar_con_mapeo(df_list, mapeo_features, flags_kepler):
    """Unifica datasets mapeando columnas entre misiones"""

    dfs_unificados = []

    for df in df_list:
        origen = df['origen'].iloc[0]
        clase = df['clase'].iloc[0]

        print(f"\nProcesando: {origen} - {clase}")

        # Crear dataframe con columnas normalizadas
        df_normalizado = pd.DataFrame()

        # Mapear features principales
        features_encontradas = 0
        for feature_nombre, mapeo in mapeo_features.items():
            col_original = mapeo.get(origen)

            if col_original and col_original in df.columns:
                df_normalizado[feature_nombre] = df[col_original]
                features_encontradas += 1

        # Agregar flags de Kepler si es Kepler
        if origen == 'Kepler':
            for flag_nombre, col_original in flags_kepler.items():
                if col_original in df.columns:
                    df_normalizado[flag_nombre] = df[col_original]
                    features_encontradas += 1

        # Agregar clase y origen
        df_normalizado['clase'] = clase
        df_normalizado['origen'] = origen

        print(f"  Features mapeadas: {features_encontradas}")

        dfs_unificados.append(df_normalizado)

    # Combinar todos
    df_combined = pd.concat(dfs_unificados, ignore_index=True)

    return df_combined

df_combined = unificar_con_mapeo(df_list, MAPEO_FEATURES, FLAGS_KEPLER)

print(f"\n{'='*70}")
print(f"DATASET UNIFICADO CREADO")
print(f"{'='*70}")
print(f"Total filas: {len(df_combined)}")
print(f"Total columnas: {len(df_combined.columns)}")
print(f"\nDistribución de clases:")
print(df_combined['clase'].value_counts())
```

```python
# 5. ANÁLISIS DE VALORES FALTANTES

print(f"\n{'='*70}")
print("ANÁLISIS DE VALORES FALTANTES")
print(f"{'='*70}")

missing = df_combined.drop(['clase', 'origen'], axis=1).isnull().sum()
missing_pct = (missing / len(df_combined) * 100).round(2)
missing_df = pd.DataFrame({
    'Feature': missing.index,
    'Faltantes': missing.values,
    'Porcentaje': missing_pct.values
}).sort_values('Porcentaje', ascending=False)

print(missing_df.to_string(index=False))

# Eliminar columnas con más del 70% de valores faltantes
umbral = 70
cols_eliminar = missing_df[missing_df['Porcentaje'] > umbral]['Feature'].tolist()

if len(cols_eliminar) > 0:
    print(f"\n⚠ Eliminando {len(cols_eliminar)} columnas con >{umbral}% valores faltantes:")
    for col in cols_eliminar:
        pct = missing_df[missing_df['Feature']==col]['Porcentaje'].values[0]
        print(f"  • {col} ({pct:.1f}%)")
    df_combined = df_combined.drop(columns=cols_eliminar)
    print(f"\n✓ Estas columnas tienen muy pocos datos útiles")
    print(f"  Rellenar con mediana introduce más ruido que información")
```

```python
# 6. CREACIÓN DE FEATURES DERIVADAS


print(f"\n{'='*70}")
print("CREACIÓN DE FEATURES DERIVADAS")
print(f"{'='*70}")

# Features derivadas con significado físico
features_derivadas = []

# 1. Densidad planetaria relativa (Radio/Masa si ambos están disponibles)
if 'Radio_Planeta_RE' in df_combined.columns and 'Masa_Planeta_ME' in df_combined.columns:
    df_combined['Densidad_Relativa'] = df_combined['Masa_Planeta_ME'] / (df_combined['Radio_Planeta_RE'] ** 3)
    features_derivadas.append('Densidad_Relativa')

# 2. Zona habitable (basado en insolación)
if 'Flujo_Insolacion_FE' in df_combined.columns:
    # 0.25 a 4 flujos terrestres es aproximadamente la zona habitable
    df_combined['En_Zona_Habitable'] = ((df_combined['Flujo_Insolacion_FE'] >= 0.25) &
                                         (df_combined['Flujo_Insolacion_FE'] <= 4.0)).astype(int)
    features_derivadas.append('En_Zona_Habitable')

# 3. Radio estelar vs Radio planetario (contraste)
if 'Radio_Estelar_RS' in df_combined.columns and 'Radio_Planeta_RE' in df_combined.columns:
    # Convertir a mismas unidades (Radio solar = 109 radios terrestres aprox)
    df_combined['Ratio_Radio_Planeta_Estrella'] = df_combined['Radio_Planeta_RE'] / (df_combined['Radio_Estelar_RS'] * 109)
    features_derivadas.append('Ratio_Radio_Planeta_Estrella')

# 4. Profundidad teórica del tránsito (basado en radios)
if 'Ratio_Radio_Planeta_Estrella' in df_combined.columns:
    df_combined['Profundidad_Teorica_ppm'] = (df_combined['Ratio_Radio_Planeta_Estrella'] ** 2) * 1e6
    features_derivadas.append('Profundidad_Teorica_ppm')

# 5. Discrepancia entre profundidad observada y teórica (detector de FPs)
if 'Profundidad_Transito_ppm' in df_combined.columns and 'Profundidad_Teorica_ppm' in df_combined.columns:
    df_combined['Discrepancia_Profundidad'] = np.abs(df_combined['Profundidad_Transito_ppm'] -
                                                       df_combined['Profundidad_Teorica_ppm']) / df_combined['Profundidad_Teorica_ppm']
    features_derivadas.append('Discrepancia_Profundidad')

# 6. Temperatura de equilibrio vs Temperatura efectiva estelar
if 'Temp_Equilibrio_K' in df_combined.columns and 'Teff_Estelar_K' in df_combined.columns:
    df_combined['Ratio_Temp_Planeta_Estrella'] = df_combined['Temp_Equilibrio_K'] / df_combined['Teff_Estelar_K']
    features_derivadas.append('Ratio_Temp_Planeta_Estrella')

# 7. Velocidad orbital aproximada (a partir del periodo y semieje mayor)
if 'Periodo_Orbital_dias' in df_combined.columns and 'Semieje_Mayor_AU' in df_combined.columns:
    # v = 2π * a / P  (a en AU, P en días, resultado en AU/día)
    df_combined['Velocidad_Orbital_AU_dia'] = (2 * np.pi * df_combined['Semieje_Mayor_AU']) / df_combined['Periodo_Orbital_dias']
    features_derivadas.append('Velocidad_Orbital_AU_dia')

# 8. Categoría de tamaño planetario
if 'Radio_Planeta_RE' in df_combined.columns:
    def categorizar_planeta(radio):
        if pd.isna(radio):
            return np.nan
        elif radio < 1.25:
            return 1  # Tamaño Tierra
        elif radio < 2.0:
            return 2  # Super-Tierra
        elif radio < 6.0:
            return 3  # Neptuno
        else:
            return 4  # Júpiter

    df_combined['Categoria_Tamano'] = df_combined['Radio_Planeta_RE'].apply(categorizar_planeta)
    features_derivadas.append('Categoria_Tamano')

# 9. Score total de flags de Kepler (suma de flags de FP)
flags_fp = ['Flag_No_Transito', 'Flag_Eclipse_Estelar', 'Flag_Offset_Centroide', 'Flag_Contaminacion']
flags_presentes = [f for f in flags_fp if f in df_combined.columns]
if len(flags_presentes) > 0:
    df_combined['Score_Flags_FP'] = df_combined[flags_presentes].sum(axis=1)
    features_derivadas.append('Score_Flags_FP')

print(f"✓ Features derivadas creadas: {len(features_derivadas)}")
for feat in features_derivadas:
    print(f"  • {feat}")
```

```python
# 7. CONFIGURACIÓN

print(f"\n{'='*70}")
print("CONFIGURACIÓN DEL MODELO")
print(f"{'='*70}")

# CONFIGURACIÓN FIJA: 2 clases, candidatos aparte
INCLUIR_CANDIDATOS = False
print("\n✓ CONFIGURACIÓN: 2 clases (CONFIRMADO vs FALSO_POSITIVO)")
print("\n  Estrategia:")
print("    1. Entrenar con datos confirmados (verdad conocida)")
print("    2. Mantener candidatos separados para evaluación")
print("    3. Después del entrenamiento, clasificar candidatos")
print("\n  El modelo aprenderá a distinguir:")
print("    • CONFIRMADO: Exoplaneta real")
print("    • FALSO_POSITIVO: NO es exoplaneta")
print("\n  Los CANDIDATOS se clasificarán al final usando el modelo entrenado")

clases_usar = ['CONFIRMADO', 'FALSO_POSITIVO']

# Filtrar clases
df_train = df_combined[df_combined['clase'].isin(clases_usar)].copy()

print(f"\n✓ Datos para entrenamiento: {len(df_train)} filas")
print(df_train['clase'].value_counts())

# Visualización
plt.figure(figsize=(10, 6))
clase_counts = df_train['clase'].value_counts()
clase_counts.plot(kind='bar', color=['#2ecc71', '#e74c3c', '#3498db'][:len(clase_counts)])
plt.title('Distribución de Clases para Entrenamiento', fontsize=14, fontweight='bold')
plt.xlabel('Clase')
plt.ylabel('Cantidad')
plt.xticks(rotation=45)
for i, v in enumerate(clase_counts):
    plt.text(i, v + 50, str(v), ha='center', va='bottom', fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# 8. PREPROCESAMIENTO

print(f"\n{'='*70}")
print("PREPROCESAMIENTO")
print(f"{'='*70}")

# Separar features y target
X = df_train.drop(['clase', 'origen'], axis=1)
y = df_train['clase']

print(f"\nFeatures totales: {X.shape[1]}")
print(f"Muestras: {X.shape[0]}")

# Codificar
le = LabelEncoder()
y_encoded = le.fit_transform(y)

print(f"\n✓ Codificación:")
for i, clase in enumerate(le.classes_):
    cantidad = sum(y_encoded == i)
    print(f"  {clase} → {i} ({cantidad} muestras, {cantidad/len(y)*100:.1f}%)")

# Imputar y normalizar
imputer = SimpleImputer(strategy='median')
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_imputed)

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded
)

print(f"\n✓ Train: {len(X_train)} | Test: {len(X_test)}")
```

```python
# 9. ENTRENAMIENTO

print(f"\n{'='*70}")
print("ENTRENAMIENTO")
print(f"{'='*70}")

# Random Forest
print("\n[1/2] Random Forest...")
rf_model = RandomForestClassifier(
    n_estimators=300,
    max_depth=30,
    min_samples_split=10,
    min_samples_leaf=4,
    max_features='sqrt',
    class_weight='balanced',
    random_state=42,
    n_jobs=-1
)

rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
y_proba_rf = rf_model.predict_proba(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
print(f"✓ Accuracy: {acc_rf:.4f}")

cv_scores_rf = cross_val_score(rf_model, X_train, y_train, cv=5)
print(f"✓ CV Accuracy: {cv_scores_rf.mean():.4f} (+/- {cv_scores_rf.std()*2:.4f})")

# Gradient Boosting
print("\n[2/2] Gradient Boosting...")
gb_model = GradientBoostingClassifier(
    n_estimators=200,
    learning_rate=0.05,
    max_depth=7,
    min_samples_split=10,
    subsample=0.8,
    random_state=42
)

gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)
y_proba_gb = gb_model.predict_proba(X_test)
acc_gb = accuracy_score(y_test, y_pred_gb)
print(f"✓ Accuracy: {acc_gb:.4f}")

cv_scores_gb = cross_val_score(gb_model, X_train, y_train, cv=5)
print(f"✓ CV Accuracy: {cv_scores_gb.mean():.4f} (+/- {cv_scores_gb.std()*2:.4f})")
```

```python
# 10. EVALUACIÓN

mejor_modelo = rf_model if acc_rf > acc_gb else gb_model
mejor_nombre = "Random Forest" if acc_rf > acc_gb else "Gradient Boosting"
mejor_pred = y_pred_rf if acc_rf > acc_gb else y_pred_gb
mejor_proba = y_proba_rf if acc_rf > acc_gb else y_proba_gb

print(f"\n{'='*70}")
print(f"MEJOR MODELO: {mejor_nombre}")
print(f"{'='*70}")
print(f"Accuracy: {max(acc_rf, acc_gb):.4f}")

print(f"\n{classification_report(y_test, mejor_pred, target_names=le.classes_, digits=4)}")

# Matriz de confusión
cm = confusion_matrix(y_test, mejor_pred)
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu',
            xticklabels=le.classes_, yticklabels=le.classes_)
plt.title(f'{mejor_nombre} - Accuracy: {max(acc_rf, acc_gb):.4f}', fontweight='bold', fontsize=14)
plt.ylabel('Real')
plt.xlabel('Predicha')
plt.tight_layout()
plt.show()

# Feature Importance
feature_importance = pd.DataFrame({
    'feature': X.columns,
    'importance': mejor_modelo.feature_importances_
}).sort_values('importance', ascending=False)

top_n = min(20, len(feature_importance))
plt.figure(figsize=(12, 8))
plt.barh(range(top_n), feature_importance['importance'].head(top_n), color='steelblue')
plt.yticks(range(top_n), feature_importance['feature'].head(top_n))
plt.xlabel('Importancia')
plt.title(f'Top {top_n} Features - {mejor_nombre}', fontweight='bold')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()

print(f"\nTop 15 Features:")
print(feature_importance.head(15).to_string(index=False))
```

```python
# 11. EVALUACIÓN DE CANDIDATOS

print(f"\n{'='*70}")
print("EVALUACIÓN DE CANDIDATOS (NO USADOS EN ENTRENAMIENTO)")
print(f"{'='*70}")

# Obtener todos los candidatos que fueron separados
candidatos = df_combined[df_combined['clase'] == 'CANDIDATO'].copy()

if len(candidatos) > 0:
    print(f"\n✓ Total de candidatos a evaluar: {len(candidatos)}")
    print(f"\nEstos objetos NO fueron usados en el entrenamiento.")
    print(f"El modelo los clasificará basándose en lo que aprendió de")
    print(f"confirmados y falsos positivos.\n")

    # Preparar features de candidatos
    X_cand = candidatos.drop(['clase', 'origen'], axis=1)
    X_cand_imp = imputer.transform(X_cand)
    X_cand_scaled = scaler.transform(X_cand_imp)

    # Predecir con el mejor modelo
    pred_cand = mejor_modelo.predict(X_cand_scaled)
    proba_cand = mejor_modelo.predict_proba(X_cand_scaled)

    # Crear dataframe de resultados
    resultados_candidatos = pd.DataFrame({
        'Origen': candidatos['origen'].values,
        'Predicción': le.inverse_transform(pred_cand),
        'Prob_Confirmado': proba_cand[:, 0] * 100,
        'Prob_FalsoPositivo': proba_cand[:, 1] * 100,
        'Confianza': proba_cand.max(axis=1) * 100
    })

    print(f"{'─'*70}")
    print("DISTRIBUCIÓN DE PREDICCIONES PARA CANDIDATOS")
    print(f"{'─'*70}")
    print(resultados_candidatos['Predicción'].value_counts())
    print()

    # Categorizar candidatos por nivel de confianza
    print(f"{'─'*70}")
    print("CATEGORIZACIÓN POR NIVEL DE CONFIANZA")
    print(f"{'─'*70}")

    confirmados_fuertes = resultados_candidatos[
        (resultados_candidatos['Predicción'] == 'CONFIRMADO') &
        (resultados_candidatos['Prob_Confirmado'] > 90)
    ]
    confirmados_moderados = resultados_candidatos[
        (resultados_candidatos['Predicción'] == 'CONFIRMADO') &
        (resultados_candidatos['Prob_Confirmado'].between(70, 90))
    ]
    confirmados_debiles = resultados_candidatos[
        (resultados_candidatos['Predicción'] == 'CONFIRMADO') &
        (resultados_candidatos['Prob_Confirmado'] < 70)
    ]
    fps_fuertes = resultados_candidatos[
        (resultados_candidatos['Predicción'] == 'FALSO_POSITIVO') &
        (resultados_candidatos['Prob_FalsoPositivo'] > 90)
    ]
    fps_moderados = resultados_candidatos[
        (resultados_candidatos['Predicción'] == 'FALSO_POSITIVO') &
        (resultados_candidatos['Prob_FalsoPositivo'].between(70, 90))
    ]
    fps_debiles = resultados_candidatos[
        (resultados_candidatos['Predicción'] == 'FALSO_POSITIVO') &
        (resultados_candidatos['Prob_FalsoPositivo'] < 70)
    ]

    print(f"\n🟢 Candidatos FUERTES para confirmación (Prob > 90%):")
    print(f"   {len(confirmados_fuertes)} candidatos")
    print(f"   → Prioridad ALTA para seguimiento")

    print(f"\n🟡 Candidatos MODERADOS para confirmación (Prob 70-90%):")
    print(f"   {len(confirmados_moderados)} candidatos")
    print(f"   → Prioridad MEDIA para seguimiento")

    print(f"\n🟠 Candidatos DÉBILES para confirmación (Prob < 70%):")
    print(f"   {len(confirmados_debiles)} candidatos")
    print(f"   → Prioridad BAJA, revisar features")

    print(f"\n🔴 Probables FALSOS POSITIVOS (Prob FP > 70%):")
    print(f"   {len(fps_fuertes) + len(fps_moderados)} candidatos")
    print(f"   → Considerar descartar o investigar anomalías")

    print(f"\n⚪ Zona AMBIGUA (ambas prob < 70%):")
    print(f"   {len(fps_debiles)} candidatos")
    print(f"   → Requieren más datos observacionales")

    # Mostrar top 10 candidatos más prometedores
    print(f"\n{'='*70}")
    print("TOP 10 CANDIDATOS MÁS PROMETEDORES")
    print(f"{'='*70}")
    print("\nCandidatos con mayor probabilidad de ser confirmados:\n")

    top_candidatos = resultados_candidatos.nlargest(10, 'Prob_Confirmado')
    print(top_candidatos.to_string(index=False))

    # Mostrar top 10 probables falsos positivos
    print(f"\n{'='*70}")
    print("TOP 10 PROBABLES FALSOS POSITIVOS")
    print(f"{'='*70}")
    print("\nCandidatos con mayor probabilidad de ser falsos positivos:\n")

    top_fps = resultados_candidatos.nlargest(10, 'Prob_FalsoPositivo')
    print(top_fps.to_string(index=False))

    # Visualización de distribución
    fig, axes = plt.subplots(1, 2, figsize=(16, 6))

    # Gráfico 1: Distribución de predicciones
    pred_counts = resultados_candidatos['Predicción'].value_counts()
    axes[0].bar(pred_counts.index, pred_counts.values, color=['#2ecc71', '#e74c3c'])
    axes[0].set_title('Clasificación de Candidatos', fontweight='bold', fontsize=14)
    axes[0].set_xlabel('Predicción del Modelo')
    axes[0].set_ylabel('Cantidad de Candidatos')
    for i, v in enumerate(pred_counts.values):
        axes[0].text(i, v + 50, str(v), ha='center', va='bottom', fontweight='bold')

    # Gráfico 2: Histograma de probabilidades de confirmación
    axes[1].hist(resultados_candidatos['Prob_Confirmado'], bins=20,
                 color='steelblue', edgecolor='black', alpha=0.7)
    axes[1].axvline(x=50, color='orange', linestyle='--', linewidth=2, label='50% (umbral)')
    axes[1].axvline(x=90, color='green', linestyle='--', linewidth=2, label='90% (alta confianza)')
    axes[1].set_title('Distribución de Prob(Confirmado) en Candidatos',
                      fontweight='bold', fontsize=14)
    axes[1].set_xlabel('Probabilidad de ser Confirmado (%)')
    axes[1].set_ylabel('Cantidad de Candidatos')
    axes[1].legend()
    axes[1].grid(alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Guardar resultados de candidatos
    resultados_candidatos.to_csv('candidatos_clasificados.csv', index=False)
    print(f"\n✓ Resultados guardados en: 'candidatos_clasificados.csv'")
    files.download('candidatos_clasificados.csv')

else:
    print("\n⚠ No hay candidatos para evaluar")
```

```python
# 12. ANÁLISIS DETALLADO: 10 EJEMPLOS


print(f"\n{'='*70}")
print("ANÁLISIS DETALLADO: 10 EJEMPLOS DE PREDICCIÓN")
print(f"{'='*70}")

print(f"\nNOTA IMPORTANTE sobre las probabilidades:")
print(f"━" * 70)
print(f"El modelo clasifica en 2 clases:")
print(f"  • CONFIRMADO: Es un exoplaneta real")
print(f"  • FALSO_POSITIVO: NO es un exoplaneta")
print(f"\nLas 2 probabilidades suman 100%:")
print(f"  Prob(CONFIRMADO) + Prob(FALSO_POSITIVO) = 100%")
print(f"\nCuando evaluemos candidatos:")
print(f"  • Si Prob(CONFIRMADO) > 90%: Candidato fuerte para confirmación")
print(f"  • Si Prob(CONFIRMADO) 50-90%: Candidato moderado, necesita más datos")
print(f"  • Si Prob(CONFIRMADO) < 50%: Probablemente falso positivo")
print(f"━" * 70)

# Seleccionar 10 ejemplos: mezclados de las 3 clases
ejemplos_por_clase = 3
indices_por_clase = {}

for clase_idx, clase_nombre in enumerate(le.classes_):
    indices_clase = np.where(y_test == clase_idx)[0]
    if len(indices_clase) >= ejemplos_por_clase:
        indices_seleccionados = np.random.choice(indices_clase, ejemplos_por_clase, replace=False)
        indices_por_clase[clase_nombre] = indices_seleccionados

# Combinar todos los índices
todos_indices = []
for indices in indices_por_clase.values():
    todos_indices.extend(indices)
todos_indices = todos_indices[:10]  # Tomar solo 10

print(f"\n{'='*70}")
print(f"Analizando {len(todos_indices)} ejemplos del conjunto de prueba")
print(f"{'='*70}\n")

resultados_detallados = []

for i, idx in enumerate(todos_indices, 1):
    clase_real = le.inverse_transform([y_test[idx]])[0]
    clase_pred = le.inverse_transform([mejor_pred[idx]])[0]
    probas = mejor_proba[idx]

    # Crear diccionario de resultados para 2 clases
    resultado = {
        'Num': i,
        'Clase_Real': clase_real,
        'Clase_Predicha': clase_pred,
        'Correcto': '✓' if clase_real == clase_pred else '✗',
        'Prob_Confirmado': probas[0] * 100,
        'Prob_FalsoPositivo': probas[1] * 100,
        'Confianza': max(probas) * 100
    }
    resultados_detallados.append(resultado)

    # Imprimir análisis individual
    print(f"{'─' * 70}")
    print(f"EJEMPLO #{i}")
    print(f"{'─' * 70}")
    print(f"🎯 Clase Real:      {clase_real}")
    print(f"🤖 Clase Predicha:  {clase_pred}")
    print(f"{'✓ CORRECTO' if clase_real == clase_pred else '✗ INCORRECTO'}")
    print(f"\n📊 Probabilidades del modelo:")

    for j, clase in enumerate(le.classes_):
        prob = probas[j] * 100
        es_prediccion = '← PREDICCIÓN' if j == np.argmax(probas) else ''
        bar = '█' * int(prob / 2)
        print(f"   {clase:18s}: {prob:6.2f}% {bar} {es_prediccion}")

    print(f"\n💡 Confianza de la predicción: {max(probas)*100:.2f}%")

    # Mostrar algunos valores de features importantes
    ejemplo_features = X_test[idx]
    print(f"\n🔬 Features clave (primeras 5):")
    for k, col in enumerate(X.columns[:5]):
        valor_norm = ejemplo_features[k]
        # Desescalar para mostrar valor real
        valor_real = valor_norm * scaler.scale_[k] + scaler.mean_[k]
        print(f"   {col:30s}: {valor_real:.4f}")
    print()
```
